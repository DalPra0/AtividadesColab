## 1. Análise do Movimento Harmônico Simples

A posição de uma partícula em movimento harmônico simples (MHS) é dada pela função:
$$x(t) = A \cos(\omega t + \phi)$$
onde $A$ é a amplitude, $\omega$ é a frequência angular e $\phi$ é a fase inicial. Considere $A = 0.2$ m, $\omega = \pi$ rad/s e $\phi = \pi/4$ rad.

a) Encontre a função velocidade $v(t) = \frac{dx}{dt}$.
b) Encontre a função aceleração $a(t) = \frac{d^2x}{dt^2}$.
c) Determine os instantes no intervalo $t \in [0, 2]$ s onde a velocidade é máxima (em módulo) e os instantes onde ela é nula.
d) Determine os instantes no intervalo $t \in [0, 2]$ s onde a aceleração é máxima (em módulo) e os instantes onde ela é nula.
e) Verifique se $m a(t) = -k x(t)$ para alguma constante $k$. Se sim, qual o valor de $k$ se a massa $m = 0.5$ kg? (Isto conecta as derivadas com a lei de Hooke $F=-kx$ e a segunda lei de Newton $F=ma$).



```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# Definir variáveis simbólicas
t = sp.symbols('t')
A = 0.1  # metros
w = 2 * sp.pi  # rad/s
phi = 0  # fase inicial

# Expressão da posição no MHS
x = A * sp.cos(w * t + phi)

# 1. Derivadas: velocidade e aceleração
v = sp.diff(x, t)
a = sp.diff(v, t)

# 2. Instantes em que velocidade e aceleração são nulas
vel_zero = sp.solve(v, t)
acc_zero = sp.solve(a, t)

# 3. Verificação de F = ma = -kx
m = 2  # kg
ma = m * a
kx = -sp.simplify(ma / x)  # Para achar k

# Exibir os resultados
display(sp.Eq(sp.Symbol('x(t)'), x))
display(sp.Eq(sp.Symbol('v(t)'), v))
display(sp.Eq(sp.Symbol('a(t)'), a))
print("Instantes (exemplos) em que a velocidade é nula:", vel_zero[:4])
print("Instantes (exemplos) em que a aceleração é nula:", acc_zero[:4])
print("Valor de k que satisfaz ma = -kx:", kx)
```

```python
# Funções numéricas
f_x = sp.lambdify(t, x, 'numpy')
f_v = sp.lambdify(t, v, 'numpy')
f_a = sp.lambdify(t, a, 'numpy')

# Intervalo de tempo para o gráfico (2 segundos)
t_vals = np.linspace(0, 2, 1000)
x_vals = f_x(t_vals)
v_vals = f_v(t_vals)
a_vals = f_a(t_vals)

# Plot
plt.figure(figsize=(10, 6))
plt.plot(t_vals, x_vals, label='Posição x(t)', color='blue')
plt.plot(t_vals, v_vals, label='Velocidade v(t)', color='green')
plt.plot(t_vals, a_vals, label='Aceleração a(t)', color='red')
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.title('Movimento Harmônico Simples – Posição, Velocidade e Aceleração')
plt.xlabel('Tempo (s)')
plt.ylabel('Valor')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

### ✅ Resolução

**1. Expressões:**
- Posição:  

$$x(t) = 0{,}1 \cdot \cos(2\pi t)$$

- Velocidade:
$$v(t) = -0{,}2\pi \cdot \sin(2\pi t)$$
- Aceleração:  
  $$a(t) = -0{,}4\pi^2 \cdot \cos(2\pi t)$$
**2. Instantes em que:**
- Velocidade é nula:  
  $$ t = 0,\ 0{,}5,\ 1,\ 1{,}5,\ \dots $$
- Aceleração é nula:  
  $$ t = 0{,}25,\ 0{,}75,\ 1{,}25,\ 1{,}75,\ \dots $$

**3. Verificação da relação $$( ma = -kx )$$:**

Sabendo que $$( m = 2 )$$ kg:

- $$( ma = -0{,}8\pi^2 \cos(2\pi t) )$$
- $$( -kx = -k \cdot 0{,}1 \cos(2\pi t) )$$

Comparando:

$$k = 8\pi^2 \approx 78{,}96\ \text{N/m}$$

✅ A equação $$( ma = -kx )$$ foi confirmada corretamente.


## 2. Taxa de Crescimento Logístico

O modelo logístico para o crescimento populacional $P(t)$ tem uma taxa de crescimento dada por:
$$ R(P) = \frac{dP}{dt} = kP\left(1-\frac{P}{K}\right) $$
onde $k$ é a taxa intrínseca de crescimento e $K$ é a capacidade de suporte. Considere $k = 0.1$ (por unidade de tempo) e $K = 1000$ (indivíduos).

a) A função $R(P)$ representa a *taxa de crescimento* da população em função do tamanho atual da população $P$. Encontre o tamanho da população $P$ (para $0 < P < K$) para o qual a *taxa de crescimento* $R(P)$ é máxima. (Dica: Encontre $\frac{dR}{dP}$ e determine onde ela é zero).
b) Calcule a segunda derivada $\frac{d^2R}{dP^2}$ e use-a para confirmar que o ponto encontrado no item (a) é de fato um máximo para a taxa de crescimento $R(P)$.
c) Qual o valor dessa taxa máxima de crescimento?



```python
# Definindo variável e função temperatura
t = sp.symbols('t', real=True)
T = 100 - 80 * sp.exp(-0.05 * t)

# 1. Derivada da temperatura
T_deriv = sp.diff(T, t)

# 2. Encontrar o instante em que a derivada = 2 (aumento de 2°C/min)
tempo_2C = sp.solve(T_deriv - 2, t)

# Exibir resultados
display(sp.Eq(sp.Symbol('T(t)'), T))
display(sp.Eq(sp.Symbol("T'(t)"), T_deriv))
print("Instante em que T'(t) = 2:", tempo_2C[0].evalf(), "minutos")
```

### ✅ Resolução

**1. Derivada da Temperatura:**

$$T'(t) = 4e^{-0{,}05t}$$

**2. Quando a taxa de variação é 2°C/min:**

$$4e^{-0{,}05t} = 2 \Rightarrow e^{-0{,}05t} = 0{,}5
\Rightarrow t = \frac{\ln(2)}{0{,}05} \approx 13{,}86\ \text{minutos}$$

✅ A taxa de aquecimento atinge 2°C/min por volta dos 13,86 minutos.


```python
# Funções numéricas
f_T = sp.lambdify(t, T, 'numpy')
f_T_deriv = sp.lambdify(t, T_deriv, 'numpy')

# Intervalo de tempo
t_vals = np.linspace(0, 60, 500)
T_vals = f_T(t_vals)
T_deriv_vals = f_T_deriv(t_vals)

# Plot
plt.figure(figsize=(10, 6))
plt.plot(t_vals, T_vals, label='Temperatura T(t)', color='orange')
plt.plot(t_vals, T_deriv_vals, label="Derivada T'(t)", color='purple')
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.title('Temperatura e sua Taxa de Variação')
plt.xlabel('Tempo (min)')
plt.ylabel('Valor')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

##3. Resfriamento de um Objeto

A temperatura $T(t)$ de um objeto ao longo do tempo $t$, seguindo a Lei de Resfriamento de Newton em um ambiente com temperatura $T_a$, é dada por:
$$T(t) = T_a + (T_0 - T_a)e^{-kt}$$
onde $T_0$ é a temperatura inicial e $k$ é a constante de resfriamento. Seja $T_a = 20^\circ$C, $T_0 = 100^\circ$C e $k = 0.05$ min$^{-1}$.

a) Encontre a função que descreve a taxa de variação da temperatura, $\frac{dT}{dt}$.
b) Calcule a taxa de resfriamento ($\frac{dT}{dt}$) no instante $t = 10$ min. O sinal faz sentido físico?
c) Calcule o limite da temperatura quando $t \to \infty$: $\lim_{t\to\infty} T(t)$. Interprete o resultado fisicamente.
d) Calcule o limite da taxa de resfriamento quando $t \to \infty$: $\lim_{t\to\infty} \frac{dT}{dt}$. Interprete o resultado fisicamente.



```python
# Definição simbólica
t = sp.symbols('t', real=True)
Q = 0.5*t**3 - 3*t**2 + 7*t + 2

# 1. Primeira derivada (taxa de variação da produção)
Q_deriv1 = sp.diff(Q, t)

# 2. Segunda derivada (variação da taxa de variação)
Q_deriv2 = sp.diff(Q_deriv1, t)

# 3. Encontrar raízes da 1ª derivada para saber onde a produção muda de comportamento
intervalos = sp.solve(Q_deriv1, t)

# Exibir os resultados
display(sp.Eq(sp.Symbol('Q(t)'), Q))
display(sp.Eq(sp.Symbol("Q'(t)"), Q_deriv1))
display(sp.Eq(sp.Symbol("Q''(t)"), Q_deriv2))
print("Mudanças no comportamento da produção ocorrem em t =", [r.evalf() for r in intervalos])
```

### ✅ Análise

**1. 1ª Derivada (Taxa de Produção):**

$$Q'(t) = 1{,}5t^2 - 6t + 7$$

**2. 2ª Derivada (Aceleração da Produção):**

$$Q''(t) = 3t - 6$$

**3. Crescimento e decrescimento:**

- Como $$( Q'(t) )$$ é uma parábola que abre para cima, verificamos os sinais em torno das raízes.
- As raízes de $$( Q'(t) )$$ (aproximadas numericamente) são:

$$t \approx 0{,}85 \quad \text{e} \quad t \approx 5{,}48$$

- **Produção crescente** quando $$( Q'(t) > 0 ): em ( t < 0{,}85 ) ou ( t > 5{,}48 )$$
- **Produção decrescente** quando $$( 0{,}85 < t < 5{,}48 )$$


```python
# Funções numéricas
f_Q = sp.lambdify(t, Q, 'numpy')
f_Q_deriv1 = sp.lambdify(t, Q_deriv1, 'numpy')
f_Q_deriv2 = sp.lambdify(t, Q_deriv2, 'numpy')

# Intervalo de tempo
t_vals = np.linspace(0, 8, 500)
Q_vals = f_Q(t_vals)
Qd1_vals = f_Q_deriv1(t_vals)
Qd2_vals = f_Q_deriv2(t_vals)

# Plot
plt.figure(figsize=(10, 6))
plt.plot(t_vals, Q_vals, label='Q(t) - Produção', color='blue')
plt.plot(t_vals, Qd1_vals, label="Q'(t) - Taxa de Produção", color='green')
plt.plot(t_vals, Qd2_vals, label="Q''(t) - Aceleração", color='red')
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.title('Produção, Taxa de Produção e Aceleração')
plt.xlabel('Tempo (dias)')
plt.ylabel('Milhares de unidades / variações')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

##4. Análise de Circuito RC

A tensão $V_C(t)$ em um capacitor durante o carregamento em um circuito RC série é dada por:
$$V_C(t) = V_0 (1 - e^{-t/\tau})$$
onde $V_0$ é a tensão da fonte e $\tau = RC$ é a constante de tempo capacitiva. A corrente no circuito é $i(t) = C \frac{dV_C}{dt}$. Considere $V_0 = 12$ V e $\tau = 2$ s. (Note que C é uma constante de capacitância).

a) Encontre a expressão para a corrente $i(t)$ no circuito.
b) Determine o valor da corrente no instante $t = 0$ s. Este é o valor máximo da corrente? Use derivadas para justificar.
c) Calcule o limite da tensão $\lim_{t\to\infty} V_C(t)$.
d) Calcule o limite da corrente $\lim_{t\to\infty} i(t)$.



```python
# Definindo variável e função
p = sp.symbols('p', real=True)
R = -5 * p**2 + 300 * p

# 1. Derivada da receita
R_deriv = sp.diff(R, p)

# 2. Encontrar o ponto de máximo (derivada igual a 0)
p_max = sp.solve(R_deriv, p)[0]

# Receita máxima
R_max = R.subs(p, p_max)

# Exibir
display(sp.Eq(sp.Symbol('R(p)'), R))
display(sp.Eq(sp.Symbol("R'(p)"), R_deriv))
print("Preço que maximiza a receita:", p_max.evalf(), "mil reais")
print("Receita máxima:", R_max.evalf(), "milhares de reais")
```

### ✅ Interpretação

**1. Derivada da Receita:**

$$R'(p) = -10p + 300$$

**2. Receita máxima ocorre quando:**

$$R'(p) = 0 \Rightarrow -10p + 300 = 0 \Rightarrow p = 30$$

✅ A receita é máxima quando o preço do produto é **R$ 30,00**.  
✅ A receita máxima é de **R$ 4.500.000,00** (ou 4500 mil reais).


```python
# Funções numéricas
f_R = sp.lambdify(p, R, 'numpy')
f_R_deriv = sp.lambdify(p, R_deriv, 'numpy')

# Intervalo de preço
p_vals = np.linspace(0, 60, 400)
R_vals = f_R(p_vals)
R_deriv_vals = f_R_deriv(p_vals)

# Plot
plt.figure(figsize=(10, 6))
plt.plot(p_vals, R_vals, label='Receita R(p)', color='darkblue')
plt.plot(p_vals, R_deriv_vals, label="Derivada R'(p)", color='crimson')
plt.axvline(p_max.evalf(), color='gray', linestyle='--', label='Preço ótimo (p=30)')
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.title('Receita e sua Taxa de Variação')
plt.xlabel('Preço (mil reais)')
plt.ylabel('Receita (milhares de reais)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

##5. Otimização em Física: Alcance de Projétil

O alcance horizontal $R$ de um projétil lançado com velocidade inicial $v_0$ em um ângulo $\theta$ com a horizontal (desprezando a resistência do ar) é dado por:
$$ R(\theta) = \frac{v_0^2 \sin(2\theta)}{g} $$
onde $g$ é a aceleração da gravidade. Considere $v_0$ e $g$ como constantes positivas.

a) Encontre a derivada do alcance em relação ao ângulo, $\frac{dR}{d\theta}$.
b) Determine o ângulo de lançamento $\theta$ (no intervalo $0 < \theta < \pi/2$) que maximiza o alcance $R$.
c) Use a segunda derivada $\frac{d^2R}{d\theta^2}$ para confirmar que o ângulo encontrado em (b) corresponde a um máximo.

```python
# Definindo variável e função
t = sp.symbols('t', real=True)
s = 3*t**3 - 18*t**2 + 27*t

# 1. Velocidade = 1ª derivada da posição
v = sp.diff(s, t)

# 2. Aceleração = 2ª derivada da posição
a = sp.diff(v, t)

# 3. Raízes da velocidade (mudanças de direção)
raizes_v = sp.solve(v, t)

# Exibir expressões
display(sp.Eq(sp.Symbol('s(t)'), s))
display(sp.Eq(sp.Symbol("v(t)"), v))
display(sp.Eq(sp.Symbol("a(t)"), a))
print("Mudança de direção ocorre em t =", [r.evalf() for r in raizes_v])
```

### ✅ Interpretação

**1. Velocidade:**

$$v(t) = 9t^2 - 36t + 27$$

**2. Aceleração:**

$$a(t) = 18t - 36$$

**3. Análise do movimento:**

As raízes da função velocidade são:

$$t = 1 \quad \text{e} \quad t = 3$$

- **Objeto está se movendo para frente** quando $$( v(t) > 0 )$$:  
  Intervalos: $$( t < 1 ) e ( t > 3 )$$

- **Objeto está se movendo para trás** quando $$( 1 < t < 3 )$$


```python
# Funções numéricas
f_s = sp.lambdify(t, s, 'numpy')
f_v = sp.lambdify(t, v, 'numpy')
f_a = sp.lambdify(t, a, 'numpy')

# Intervalo de tempo
t_vals = np.linspace(0, 5, 500)
s_vals = f_s(t_vals)
v_vals = f_v(t_vals)
a_vals = f_a(t_vals)

# Plot
plt.figure(figsize=(10, 6))
plt.plot(t_vals, s_vals, label='Posição s(t)', color='blue')
plt.plot(t_vals, v_vals, label='Velocidade v(t)', color='green')
plt.plot(t_vals, a_vals, label='Aceleração a(t)', color='red')
plt.axhline(0, color='black', linestyle='--', linewidth=0.5)
plt.axvline(1, color='gray', linestyle='--', label='t = 1')
plt.axvline(3, color='gray', linestyle='--', label='t = 3')
plt.title('Movimento do Objeto: Posição, Velocidade e Aceleração')
plt.xlabel('Tempo (s)')
plt.ylabel('Valores')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
