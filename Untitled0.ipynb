## Questão 1

Questão 1: Variáveis e Constantes

Você precisa implementar um sistema que calcula a pontuação final de um jogo. Cada jogador começa com 100 pontos. A cada rodada, eles podem ganhar ou perder pontos. A pontuação máxima permitida é de 1000, e a mínima é 0. Além disso, uma taxa fixa de 10% é aplicada ao saldo final como taxa de serviço.

Desafio:

    Escreva um código que mantém as regras acima utilizando variáveis e constantes adequadamente.
    Explique por que a escolha de determinadas variáveis como constantes ou variáveis mutáveis é importante no contexto deste sistema.

````python
def calcular_pontuacao_final():
    rodada = 1
    jogador1 = 100
    jogador2 = 100
    taxa = 0.10
    vencedor = ""
    pontosVencedor = 0

    while jogador1 < 1000 and jogador2 < 1000:
        print(f"Rodada {rodada}")
        pontos = int(input("Quem fez pontos nessa rodada? Jogador1 = 1, Jogador2 = 2 "))

        if pontos == 1:
            quantPontos = int(input("Quantos pontos o jogador 1 fez? "))
            jogador1 += quantPontos
            if jogador1 > 1000:
                jogador1 = 1000  # Limita a pontuação ao máximo de 1000
            elif jogador1 < 0:
                jogador1 = 0  # Garante que a pontuação não seja menor que 0
        elif pontos == 2:
            quantPontos = int(input("Quantos pontos o jogador 2 fez? "))
            jogador2 += quantPontos
            if jogador2 > 1000:
                jogador2 = 1000  # Limita a pontuação ao máximo de 1000
            elif jogador2 < 0:
                jogador2 = 0  # Garante que a pontuação não seja menor que 0
        else:
            print("Jogador inexistente! Digite novamente.")
            continue  # Volta para o início do loop sem incrementar a rodada

        rodada += 1

    if jogador1 > jogador2:
        vencedor = "Jogador 1"
        pontosVencedor = jogador1
    else:
        vencedor = "Jogador 2"
        pontosVencedor = jogador2

    pontosVencedorFinal = pontosVencedor * (1 - taxa)
    print(f"O vencedor foi {vencedor} com {pontosVencedor} pontos antes da taxa e {pontosVencedorFinal:.2f} pontos após a aplicação da taxa de serviço!")

calcular_pontuacao_final()
````

Questão 2: Estruturas Condicionais

Considere um sistema de autenticação que valida um usuário baseado em três níveis de segurança: senha, reconhecimento facial, e autenticação de dois fatores (2FA). Se a senha estiver incorreta, o sistema deve negar o acesso imediatamente. Se a senha estiver correta, mas o reconhecimento facial falhar, o sistema deve solicitar o 2FA. Se a senha e o reconhecimento facial forem bem-sucedidos, o acesso é concedido sem solicitar o 2FA.

Desafio:

    Implemente o código que simule esse sistema de autenticação.
    Explique como a ordem das condições afeta a segurança e eficiência do sistema.

````python
def sistema_autenticacao(senha_correta, senha_input, reconhecimento_facial, dois_fatores):
    if senha_input != senha_correta:
        print("Acesso negado: senha incorreta.")
        return False
    elif not reconhecimento_facial:
        if dois_fatores:
            print("Acesso concedido após autenticação de dois fatores.")
            return True
        else:
            print("Acesso negado: reconhecimento facial e 2FA falharam.")
            return False
    else:
        print("Acesso concedido sem necessidade de 2FA.")
        return True

# Exemplo de uso
sistema_autenticacao("senha123", "senha123", True, False)
````

Questão 3: Repetições e Loops

Você foi encarregado de desenvolver um algoritmo para processar uma grande lista de números inteiros e determinar quantos números são primos. Contudo, o desafio é fazer isso com o menor número possível de iterações, considerando que a lista pode conter milhões de números.

Desafio:

    Escreva um código otimizado que encontre e conte todos os números primos na lista.
    Justifique como você minimizou as iterações necessárias, considerando técnicas como a crivagem ou outros algoritmos avançados.

````python
def contar_primos(lista_numeros):
    def eh_primo(n):
        if n < 2:
            return False
        for i in range(2, int(n ** 0.5) + 1):
            if n % i == 0:
                return False
        return True

    total_primos = 0
    for numero in lista_numeros:
        if eh_primo(numero):
            total_primos += 1

    return total_primos

# Exemplo de uso
numeros = [2, 3, 4, 5, 10, 13, 17, 19, 23, 24, 29]
print(contar_primos(numeros))
````

## Questão 4

Questão 4: Manipulação de Strings

Em um sistema de gerenciamento de conteúdo, você precisa criar uma função que sanitiza títulos de artigos, removendo todos os caracteres especiais e substituindo espaços por hífens. Além disso, a primeira letra de cada palavra deve ser maiúscula, exceto por palavras de ligação como "e", "de", "da", "do", etc.

Desafio:

    Implemente a função sanitize_title(titulo) que realiza todas as operações acima.
    Explique como o uso de métodos de string avançados e compreensão de listas pode tornar seu código mais eficiente.

````python
def sanitize_title(titulo):
    palavras_excluidas = ["e", "de", "da", "do", "das", "dos"]
    titulo_sanitizado = ""

    for char in titulo:
        if char.isalnum() or char.isspace():
            titulo_sanitizado += char

    palavras = titulo_sanitizado.split()
    titulo_final = "-".join([palavra.capitalize() if palavra.lower() not in palavras_excluidas else palavra.lower() for palavra in palavras])

    return titulo_final

# Exemplo de uso
titulo = "O estranho caso de Dr. Jekyll e Mr. Hyde!"
print(sanitize_title(titulo))
````

## 1. **Variáveis**

### O que são Variáveis?

- **Definição:** Em programação, variáveis são locais na memória do computador onde podemos armazenar dados que podem ser alterados durante a execução do programa. Elas são fundamentais para manipular e processar informações.

### Como Declarar Variáveis em Python?

- **Sintaxe:** Em Python, as variáveis são declaradas simplesmente atribuindo um valor a um nome. Não é necessário declarar o tipo da variável (Python é uma linguagem de tipagem dinâmica).
  
  ```python
  x = 10  # x é uma variável que contém o valor 10
  nome = "João"  # nome é uma variável que contém a string "João"
  ```

### Convenções de Nomenclatura

- **Identificadores:** O nome de uma variável deve começar com uma letra (a-z, A-Z) ou um sublinhado (_), seguido de letras, números ou sublinhados. Python é sensível a maiúsculas e minúsculas.
  
  ```python
  _contador = 1
  taxaJuros = 0.05
  ```

- **Palavras Reservadas:** Python possui palavras reservadas (como `if`, `while`, `return`) que não podem ser usadas como nomes de variáveis.

### Tipos de Dados e Tipagem Dinâmica

- **Tipos de Dados Comuns:**
  - **Inteiros (`int`):** Números inteiros, positivos ou negativos.
  - **Flutuantes (`float`):** Números reais com parte decimal.
  - **Strings (`str`):** Sequências de caracteres (texto).
  - **Booleanos (`bool`):** Valores `True` ou `False`.
  - **Listas (`list`), Tuplas (`tuple`), Dicionários (`dict`), Conjuntos (`set`):** Estruturas de dados compostas.

  ```python
  idade = 25  # int
  altura = 1.75  # float
  nome = "Alice"  # str
  ativo = True  # bool
  ```

- **Tipagem Dinâmica:** O tipo de uma variável em Python é determinado automaticamente no momento da atribuição, e pode ser mudado ao longo do programa.

  ```python
  x = 10  # x é um int
  x = "dez"  # agora x é uma string
  ```

### Atribuições Múltiplas e Descompactação

- **Atribuição Múltipla:** Python permite atribuir valores a múltiplas variáveis em uma única linha.

  ```python
  a, b, c = 1, 2, 3
  ```

- **Descompactação de Listas/Tuplas:** É possível atribuir os elementos de uma lista ou tupla a várias variáveis de uma vez.

  ```python
  lista = [1, 2, 3]
  x, y, z = lista
  ```

## 2. **Constantes**

### O que são Constantes?

- **Definição:** Constantes são valores que, uma vez atribuídos, não devem mudar durante a execução do programa. Elas são usadas para armazenar valores que precisam permanecer inalterados, como a taxa de juros em um cálculo financeiro ou o valor de Pi em uma equação.

### Constantes em Python

- **Convencionalmente:** Python não possui uma palavra-chave específica para declarar constantes. Em vez disso, por convenção, usamos nomes de variáveis em maiúsculas para indicar que uma variável deve ser tratada como constante.

  ```python
  PI = 3.14159
  TAXA_JUROS = 0.05
  ```

- **Imutabilidade:** Apesar da convenção, as constantes em Python podem ser alteradas, pois o Python não impede a reatribuição de valores. Cabe ao programador respeitar o uso de constantes.

### Benefícios do Uso de Constantes

- **Legibilidade:** Usar constantes ajuda a tornar o código mais legível e compreensível, pois deixa claro quais valores não devem ser modificados.
- **Manutenção:** Facilita a manutenção do código, pois se um valor constante precisar ser alterado, isso pode ser feito em um único lugar.

## 3. **Condições (Estruturas Condicionais)**

### O que são Estruturas Condicionais?

- **Definição:** Estruturas condicionais permitem que o código execute diferentes blocos de comandos com base em certas condições. São a base para a tomada de decisões em um programa.

### Estruturas Condicionais em Python

#### **`if`, `elif`, `else`**

- **`if`:** Executa um bloco de código se uma condição for verdadeira.
- **`elif`:** Executa um bloco de código se a condição anterior foi falsa, mas esta é verdadeira.
- **`else`:** Executa um bloco de código se todas as condições anteriores forem falsas.

  ```python
  idade = 20
  
  if idade >= 18:
      print("Você é maior de idade.")
  elif idade >= 13:
      print("Você é um adolescente.")
  else:
      print("Você é uma criança.")
  ```

- **Aninhamento:** Condicionais podem ser aninhadas dentro de outras, permitindo a construção de lógicas complexas.

  ```python
  if idade >= 18:
      if idade >= 65:
          print("Você é um idoso.")
      else:
          print("Você é um adulto.")
  ```

### Comparações e Operadores Lógicos

- **Operadores de Comparação:** Usados para comparar valores.
  - `==` (igual a)
  - `!=` (diferente de)
  - `>` (maior que)
  - `<` (menor que)
  - `>=` (maior ou igual a)
  - `<=` (menor ou igual a)

- **Operadores Lógicos:** Combinam múltiplas condições.
  - `and`: Verdadeiro se todas as condições forem verdadeiras.
  - `or`: Verdadeiro se pelo menos uma condição for verdadeira.
  - `not`: Inverte o valor lógico de uma condição.

  ```python
  x = 10
  if x > 5 and x < 15:
      print("x está entre 5 e 15.")
  ```

### Condicionais em Uma Linha

- **Sintaxe Condicional Ternária:** Python permite escrever condições simples em uma única linha.

  ```python
  idade = 18
  status = "maior de idade" if idade >= 18 else "menor de idade"
  print(status)
  ```

## 4. **Repetições (Loops)**

### O que são Loops?

- **Definição:** Loops permitem que o código execute repetidamente um bloco de comandos enquanto uma condição é verdadeira. Isso é útil para processar coleções de dados, realizar cálculos repetitivos, etc.

### Tipos de Loops em Python

#### **`while` Loop**

- **Definição:** Executa um bloco de código enquanto a condição especificada é verdadeira.

  ```python
  contador = 0
  while contador < 5:
      print(contador)
      contador += 1
  ```

- **Uso:** Ideal quando o número de iterações não é conhecido antecipadamente.

- **Loop Infinito:** Se a condição nunca se tornar falsa, o loop continuará indefinidamente.

  ```python
  while True:
      print("Loop infinito")
  ```

#### **`for` Loop**

- **Definição:** Itera sobre uma sequência (como uma lista, tupla, string, etc.) e executa um bloco de código para cada item.

  ```python
  lista = [1, 2, 3, 4, 5]
  for numero in lista:
      print(numero)
  ```

- **Uso:** Ideal para iterar sobre coleções de dados ou quando o número de iterações é conhecido.

- **Range:** Usado com `for` para gerar uma sequência de números.

  ```python
  for i in range(5):  # i vai de 0 a 4
      print(i)
  ```

### Controle de Fluxo em Loops

- **`break`:** Sai do loop imediatamente.

  ```python
  for i in range(10):
      if i == 5:
          break
      print(i)
  ```

- **`continue`:** Pula para a próxima iteração do loop.

  ```python
  for i in range(10):
      if i % 2 == 0:
          continue
      print(i)  # Imprime apenas números ímpares
  ```

- **`else` com Loops:** Executa um bloco de código quando o loop termina normalmente (não por `break`).

  ```python
  for i in range(5):
      print(i)
  else:
      print("Loop completo")
  ```

## 1. **Variáveis** (Variáveis, Tipagem e Escopo)

### 1.1 O que são variáveis?
- **Definição:** Em Python, uma variável é uma referência a um objeto. Diferente de linguagens como C ou Java, onde variáveis são vistas como "caixas" que contêm valores, em Python, uma variável armazena uma referência a um objeto na memória.

### 1.2 Tipagem Dinâmica e Mudança de Tipo
- **Tipagem Dinâmica:** Python é uma linguagem dinamicamente tipada, ou seja, o tipo da variável é determinado durante a execução, e não precisa ser declarado antecipadamente. Isso significa que você pode reatribuir qualquer tipo de valor a uma variável existente.

  ```python
  x = 10  # x é um int
  x = "dez"  # agora x é uma string
  ```

- **Conversão Explícita de Tipos:** Mesmo sendo dinamicamente tipada, Python permite conversões explícitas de tipos usando funções como `int()`, `float()`, `str()`, etc.

  ```python
  numero = "123"
  numero_inteiro = int(numero)  # Converte string para int
  ```

### 1.3 Atribuições e Operadores de Atribuição

- **Operadores Combinados:** Além de `=` para atribuição simples, temos operadores como `+=`, `-=`, `*=`, `/=`, que combinam operações matemáticas com atribuição.

  ```python
  x = 5
  x += 2  # Equivalente a x = x + 2
  ```

### 1.4 Escopo de Variáveis

- **Escopo Local:** Variáveis declaradas dentro de uma função ou bloco de código só existem dentro desse escopo.
  
  ```python
  def funcao():
      x = 10  # Escopo local
      print(x)
  ```

- **Escopo Global:** Variáveis declaradas fora de funções ou blocos de código têm escopo global e podem ser acessadas em qualquer parte do programa.

  ```python
  x = 5  # Variável global

  def funcao():
      print(x)  # Pode acessar a variável global
  ```

- **Palavra-chave `global`:** Se você quiser modificar uma variável global dentro de uma função, você precisa declará-la como global.

  ```python
  x = 5

  def funcao():
      global x
      x = 10  # Modifica a variável global

  funcao()
  print(x)  # Agora x vale 10
  ```

- **Palavra-chave `nonlocal`:** Em funções aninhadas, `nonlocal` permite modificar variáveis de um escopo que não seja o global.

  ```python
  def externa():
      x = 10
      def interna():
          nonlocal x
          x = 20
      interna()
      print(x)  # x vale 20
  ```

### 1.5 Boas Práticas com Variáveis

- **Nomes Descritivos:** Sempre use nomes de variáveis que sejam descritivos e indiquem claramente o que a variável armazena. Evite nomes como `x`, `y` ou `data`, a menos que o contexto seja óbvio.

  ```python
  temperatura_celsius = 36.5
  ```

## 2. **Constantes**

### 2.1 Diferença entre Constantes e Variáveis

- **Constantes:** São valores que, idealmente, não devem ser alterados durante a execução do programa. Apesar de Python não ter suporte nativo para constantes (como `const` em C++), a prática comum é usar letras maiúsculas para denotar uma constante.

  ```python
  PI = 3.14159
  ```

- **Imutabilidade:** Embora seja uma convenção tratar variáveis em maiúsculas como constantes, Python não impede que elas sejam modificadas. Cabe ao programador garantir que elas não sejam alteradas.

  ```python
  PI = 3.14159
  PI = 3  # Isso é permitido em Python, mas não recomendado.
  ```

### 2.2 Objetos Imutáveis e Mutáveis

- **Objetos Imutáveis:** Alguns tipos em Python, como `int`, `float`, `str`, e `tuple`, são imutáveis. Isso significa que, após serem criados, não podem ser alterados. Isso é importante ao trabalhar com "constantes".

- **Objetos Mutáveis:** Tipos como `list`, `dict`, e `set` são mutáveis, permitindo a modificação de seu conteúdo.

## 3. **Condições** (Condicionais)

### 3.1 Estruturas Condicionais Simples

- **`if`, `elif`, `else`:** As condições são a base da tomada de decisões em um programa. Python usa `if`, `elif` (else-if) e `else` para implementar ramificações no fluxo de execução.

  ```python
  idade = 18
  if idade >= 18:
      print("Maior de idade")
  elif 13 <= idade < 18:
      print("Adolescente")
  else:
      print("Criança")
  ```

### 3.2 Comparações e Operadores Lógicos

- **Operadores de Comparação:** São usados para comparar valores.
  - `==` (igual a)
  - `!=` (diferente de)
  - `>` (maior que)
  - `<` (menor que)
  - `>=` (maior ou igual a)
  - `<=` (menor ou igual a)

- **Operadores Lógicos:** Combinam múltiplas condições.
  - `and`: Verdadeiro se todas as condições forem verdadeiras.
  - `or`: Verdadeiro se pelo menos uma condição for verdadeira.
  - `not`: Inverte o valor lógico de uma condição.

### 3.3 Condições Complexas e Boas Práticas

- **Evitar Condições Complexas:** Condições muito complexas tornam o código difícil de entender e manter. Em vez disso, é preferível dividir as condições em variáveis intermediárias.

  ```python
  eh_maior_de_idade = idade >= 18
  tem_documento = documento_valido

  if eh_maior_de_idade and tem_documento:
      print("Pode entrar.")
  ```

- **Verificação de Valores com `in`:** O operador `in` é muito útil para verificar a presença de um valor em uma sequência.

  ```python
  vogais = "aeiou"
  letra = "e"
  
  if letra in vogais:
      print(f"{letra} é uma vogal.")
  ```

- **`is` para Comparação de Identidade:** Use o operador `is` para verificar se duas variáveis referenciam o mesmo objeto na memória (comparação de identidade, não de valor).

  ```python
  a = [1, 2, 3]
  b = a  # b referencia o mesmo objeto que a
  
  if a is b:
      print("a e b são o mesmo objeto.")
  ```

## 4. **Repetições (Loops)**

### 4.1 O Loop `for` em Python

- **Definição:** O `for` em Python é muito flexível e iterativo, usado para percorrer sequências como listas, tuplas, strings ou outros iteráveis.

  ```python
  for i in range(5):
      print(i)  # Saída: 0, 1, 2, 3, 4
  ```

- **Iterando em Sequências:**

  ```python
  lista = [10, 20, 30]
  for numero in lista:
      print(numero)
  ```

- **Enumerar Elementos:** A função `enumerate()` retorna tanto o índice quanto o valor dos itens iterados.

  ```python
  lista = ['a', 'b', 'c']
  for indice, valor in enumerate(lista):
      print(f"{indice}: {valor}")
  ```

- **Iterando sobre Dicionários:**

  ```python
  dicionario = {'nome': 'Alice', 'idade': 25}
  for chave, valor in dicionario.items():
      print(f"{chave}: {valor}")
  ```

### 4.2 O Loop `while`

- **Definição:** O `while` continua executando o bloco de código enquanto a condição permanecer verdadeira.

  ```python
  contador = 0
  while contador < 5:
      print(contador)
      contador += 1
  ```

- **Exemplo de Loop Infinito:** Um loop sem condição de término cria um loop infinito.

  ```python
  while True:
      print("Isso roda para sempre")
  ```

- **Interrompendo Loops com `

break`:** Usado para sair de um loop prematuramente.

  ```python
  while True:
      entrada = input("Digite 'sair' para encerrar: ")
      if entrada == 'sair':
          break
  ```

- **`continue`:** Pula a iteração atual e vai para a próxima.

  ```python
  for i in range(10):
      if i % 2 == 0:
          continue  # Pula números pares
      print(i)
  ```

### 4.3 `else` em Loops

- **Uso com `for` e `while`:** O bloco `else` é executado quando o loop termina naturalmente, sem interrupção via `break`.

  ```python
  for i in range(5):
      print(i)
  else:
      print("Loop terminou sem interrupção.")
  ```

### 4.4 `zip()` e Iteradores Avançados

- **`zip()` para Iteração Simultânea:** Junta duas ou mais sequências e as itera ao mesmo tempo.

  ```python
  nomes = ['Alice', 'Bob']
  idades = [25, 30]

  for nome, idade in zip(nomes, idades):
      print(f"{nome} tem {idade} anos.")
  ```
